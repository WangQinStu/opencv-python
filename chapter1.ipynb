```python
import cv2
```

```python
# 可视化图片
img = cv2.imread('Resources/lenna.png')
cv2.imshow("NameOfWidow",img)
cv2.waitKey(0)# 0代表永远等待，其它数字代表毫秒，1000ms=1s
```

```python
#可视化视频
cap = cv2.VideoCapture('Resources/video.mp4')
while True:
    ret, frame = cap.read()
    cv2.imshow("VIDEO",frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```

```python
#可视化摄像头
cap = cv2.VideoCapture(0) #将视频路径更换为摄像头id即可
cap.set(3, 640)
cap.set(4, 480)
while True:
    ret, frame = cap.read()
    cv2.imshow("VIDEO",frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```
